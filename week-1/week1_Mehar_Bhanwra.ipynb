- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    try:
        return a / b
    except TypeError:
        raise TypeError("Inputs must be numeric")
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    l = eigenvalues[i]

    left = A @ v
    right = l * v

    print(f"\nFor eigenvalue {eigenvalues[i]}:")
    print(f"LHS => A @ v = {left}")
    print(f"RHS => lambda * v = {right}")

    assert np.allclose(left, right)

# Plot
fig, ax = plt.subplots()
colors = ['magenta', 'blue']

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    scaled_v = eigenvalues[i] * v

    ax.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=f"lambda={eigenvalues[i]:.2f}")

ax.set_xlim(-2, 5); ax.set_ylim(-2, 5)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Vector Visualization')
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8); ax.legend()
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

- The eigenvector of a matrix is a non-zero vector that only gets scaled when multiplied by a transformation matrix A. This means that the vector does not rotate or change direction in many manner. It just gets larger or shorter, that is it only changes in terms of magnitude. The eigenvalue of a matrix is the factor by which the eigenvector gets scaled.

- We define eigenvectors and values as A @ x := lamda * x, where A is a square matrix in R(n x n), x is the eigenvector and lambda is a scalar in R and the corresponding eigenvalue.

- Stretching and squishing of a vector are two terms associated with scaling a vector i.e. they are geometric interpretations of scaling.

- Scaling is a mapping of the form y = Ax, with A = alpha * I. Here, alpha is a scalar value and I is an identity matrix.

- Scaling "stretches" x by a factor |alpha| when |alpha| > 1. Now, if 0 < |alpha| < 1, then the value of x is "squished". If alpha < 0, i.e. alpha is negative, the direction of the vector is reversed.

- The sign of alpha only affects the direction of the vector, whereas the magnitude determines how much the vector is stretched or squished. Geometrically, eigenvectors represent the special directions that remain unchanged under a transformation, and eigenvalues describe the amount of scaling along those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = (U * S) @ Vt
X_approx        = np.round(np.linalg.norm(X - X_reconstructed, 'fro'), 2)

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> *Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*

SVD has a geometric interpretation as a sequence of linear mappings or transformations. A matrix 'A' can be decomposed as
A = USVt

Here:
- Vt: represents a change of basis (rotation) from the standard basis into a set of orthogonal basis
- S: is a diagonal matrix containing singular values, which scale the transformed vectors by stretching, shrinking, or flipping them.
- U: change of basis (rotation) from the new orthogonal basis onto some other orientation, i.e., not necessarily where we started.

SVD decomposes a transformation into:
1. Rotation into an orthogonal basis,
2. Scaling along orthogonal directions,
3. Rotation into the output space.

In the context of machine learning applications, it is common to find matrices with many hundreds and thousands of rows and columns, sometimes reaching up to a million even.

Despite techniques such as Eigendecomposition and Singular Value Decomposition (SVD), which make the matrix factorization efficient to compute, such matrices may still consume very large amounts of time and computational resources.

One solution to this problem is to use a subset of the orthogonal vectors rather than the full set, in such a manner that we can obtain a reasonably good approximation of the original matrix.

PCA is used for this process, which is known as dimension reductionality. PCA/Principal Component Analysis is one of the most common dimensionality reduction techniques. PCA uses SVD (or eigendecomposition of the covariance matrix) to find the directions of maximum variance in the dataset. These directions are called principal components.

In the context of SVD:

- The rows of Vt (or equivalently the columns of V) contain the principal component directions.
- The singular values in S indicate the importance or variance captured by each principal component.
- The first principal component corresponds to the direction with the maximum variance, the second corresponds to the next highest variance orthogonal to the first, and so on.

Thus, the matrix Vt contains the directions of maximum variance used in PCA.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, cdf='norm')
ks_s2 = stats.kstest(s2, cdf='norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    breakpoints = np.histogram_bin_edges(expected, bins=bins)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _   = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts/len(expected)
    actual_pct   = actual_counts/len(actual)

    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct   = np.where(actual_pct == 0, epsilon, actual_pct)

    psi = np.sum((actual_pct -  expected_pct) * np.log(actual_pct/expected_pct))

    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=50, alpha=0.6, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=50, alpha=0.6, density=True, label='Production Distribution')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Statistical properties of the target variable, which the model is trying to predict changes over time. Ex. change in consumer behaviour, which is also what our machine learning model is trying to predict.
- **Covariate drift:** Covariate shift is the change in the distribution of one or more of the independent or input variables of the dataset. This means that the distribution of the features itself has changed. Ex. 10-15% increase in the salary of a population.
- **PSI < 0.1:** Implies no change or shift
- **PSI 0.1–0.2:** Implies slight change or shift
- **PSI > 0.2:** Implies large change or shift
- **Retraining trigger example:** A model retraining process can be triggered when the statistical difference between the training data distribution and the incoming production data distribution exceeds a predefined threshold. This indicates significant data drift and suggest that the model's performance may degrade if it isnot updated with newer data.

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red/total
p_blue  = blue/total
p_green = green/total

# Joint: P(red first AND blue second)
p_red_then_blue = p_red * (blue/(total - 1))

# Conditional: P(blue second | red first)
p_blue_given_red = blue/(total - 1)

# Independence check: are P(B|A) == P(B) ?
independent = p_red_then_blue == p_blue

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = ((p_free_given_spam * p_spam) + (p_free_given_ham * p_ham))

# Bayes
p_spam_given_free = ((p_free_given_spam * p_spam)/p_free)

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam
    evidence = ((p_word_given_spam * prior_spam) + (p_word_given_ham * p_ham))
    posterior = ((p_word_given_spam * p_spam)/evidence)

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

| Likelihood | P(word \| Spam) | 0.80 |
| Evidence | P(word) | 0.2750 |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([
    np.random.choice(population, size=sample_size).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std/np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.6, label='Sample Means')
x = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(x, stats.norm.pdf(x, loc=pop_mean, scale=clt_std), label='Theoretical Normal')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

The Central Limit Theorem (CLT) is important because it shows that the distribution of sample means approaches a normal distribution as the sample size increases, even in such a scenario where the original population is not normally distributed.

This allows statisticians and machine learning practitioners to use normal-distribution-based methods for estimation and hypothesis testing.

The CLT justifies the normality assumption, i.e. the assumption that the distribution of the sample means is a normal/gaussian distribution as the sample size increases regardless of the distribution of the original population, which is used in many classical statistical techniques such as t-tests, confidence intervals, linear regression inference, and ANOVA.
